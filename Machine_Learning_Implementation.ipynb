```python
d1.drop(columns=['text'],inplace=True,errors='ignore')
d1.columns
```

```python
d1
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
d1.loc[:, 'title'] = encoder.fit_transform(d1['title'])
d1.loc[:, 'text'] = encoder.fit_transform(d1['text'])
d1.loc[:, 'date'] = encoder.fit_transform(d1['date'])
d1.loc[:, 'source'] = encoder.fit_transform(d1['source'])
d1.loc[:, 'author'] = encoder.fit_transform(d1['author'])
d1.loc[:, 'category'] = encoder.fit_transform(d1['category'])
d1.loc[:, 'label'] = encoder.fit_transform(d1['label'])
d1
```

```python
X=d1.drop('category',errors='ignore')
Y=d1['category'].astype(int)
X
```

```python
Y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42,stratify=Y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,Y_train)
predict=knn.predict(X_test)
predict
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,predict)
```

```python
from sklearn.neural_network import MLPClassifier
clf=MLPClassifier(solver='lbfgs',alpha=1e-5,max_iter=2000,random_state=1)
clf.fit(X_train,Y_train)
predict1=clf.predict(X_test)
predict1
```

```python
accuracy_score(Y_test,predict1)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(X_train,Y_train)
predict2=reg.predict(X_test)
predict2
```

```python
accuracy_score(Y_test,predict2)
```
